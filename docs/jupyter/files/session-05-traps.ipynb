### Trap 1 ‚Äî predict, then run

Many people expect: **[0, 1, 4] again**.

```python
gen = (n * n for n in range(3))
first = list(gen)
list(gen)
```

<details>
<summary><strong>What really happens ‚Äî and why</strong></summary>

Result: `[]`

A generator is one-shot: after the first full pass it is exhausted. Rebuild it, or store a list if you need it twice.

</details>

### Trap 2 ‚Äî predict, then run

Many people expect: **False ‚Äî two different objects**.

```python
from dataclasses import dataclass

@dataclass
class Pt:
    x: int

p1 = Pt(1)
p2 = Pt(1)
p1 == p2
```

<details>
<summary><strong>What really happens ‚Äî and why</strong></summary>

Result: `True`

`@dataclass` writes an `__eq__` that compares fields, so equal-valued instances are `==` (even though `p1 is p2` is False).

</details>

### Trap 3 ‚Äî predict, then run

Many people expect: **'a'**.

```python
import re
m = re.match(r'<(.+)>', '<a><b>')
m.group(1)
```

<details>
<summary><strong>What really happens ‚Äî and why</strong></summary>

Result: `'a><b'`

`+` is greedy ‚Äî it grabs as much as it can. Use `+?` for the shortest match: `r'<(.+?)>'`.

</details>

### Trap 4 ‚Äî predict, then run

Many people expect: **True ‚Äî there are digits**.

```python
import re
bool(re.match(r'\d+', 'abc123'))
```

<details>
<summary><strong>What really happens ‚Äî and why</strong></summary>

Result: `False`

`re.match` anchors at the START of the string. Use `re.search` to find a match anywhere.

</details>

### Trap 5 ‚Äî predict, then run

Many people expect: **[] ‚Äî b is a different dog**.

```python
class Dog:
    tricks = []
    def teach(self, t):
        self.tricks.append(t)

a = Dog()
b = Dog()
a.teach('sit')
b.tricks
```

<details>
<summary><strong>What really happens ‚Äî and why</strong></summary>

Result: `['sit']`

`tricks` is a CLASS variable shared by every instance. Give each its own in `__init__`: `self.tricks = []`.

</details>
