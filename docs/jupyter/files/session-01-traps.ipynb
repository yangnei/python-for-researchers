# Session 1 — Traps: predict, then run

Python's dynamic typing has sharp edges. For each cell below: **read it, predict the result out loud, then run it with Shift + Enter.** When the answer surprises you, that's the cell doing its job — expand *why* and change a value to test your new understanding.

**Tips:** press **Tab** to autocomplete a name, **Shift + Tab** for a function's help. Need a library? Run `%pip install <name>` in a cell (e.g. `%pip install pandas`) — in the browser (JupyterLite) it fetches a Pyodide build and lasts for the session.

### Trap 1 — predict, then run

Many people expect: **True — the two lists are identical**.

```python
ana_scores = [88, 91]
ben_scores = [88, 91]
ana_scores is ben_scores
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `False`

`==` compares VALUES (and `ana_scores == ben_scores` is True); `is` compares IDENTITY — whether they are the same object in memory. Ana's and Ben's lists just happen to match. Use `==` for values and keep `is` for `None`.

</details>

### Trap 2 — predict, then run

Many people expect: **['Ana', 'Ben'] — backup was taken before Cara**.

```python
roster = ['Ana', 'Ben']
backup = roster
roster.append('Cara')
backup
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `['Ana', 'Ben', 'Cara']`

`backup = roster` does not copy: both names point at one list, so appending through `roster` shows up in `backup` too. Copy with `list(roster)` or `roster[:]`.

</details>

### Trap 3 — predict, then run

Many people expect: **False — a yes/no flag isn't a number**.

```python
attended = True
attended == 1
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `True`

`bool` is a subclass of `int`: `True == 1` and `False == 0`, which is why a spreadsheet's 1/0 lines up with Python's True/False.

</details>

### Trap 4 — predict, then run

Many people expect: **TypeError — you can't add a flag to a number**.

```python
base_score = 5
bonus = True
base_score + bonus
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `6`

`True` acts as `1` (and `False` as `0`) in arithmetic — a quick way to add a bonus point for a flag.

</details>

<details>

Summing booleans counts the `True`s — a handy way to count how many students passed.

### Trap 6 — predict, then run

Many people expect: **0.3**.

```python
reading_a = 0.1
reading_b = 0.2
reading_a + reading_b
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `0.30000000000000004`

Floats are stored in binary, where 0.1 and 0.2 have no exact representation, so the tiny errors add up.

</details>

### Trap 7 — predict, then run

Many people expect: **True**.

```python
total = 0.1 + 0.2
total == 0.3
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `False`

Because `0.1 + 0.2` is `0.30000000000000004`. Never test measured values with `==`; use `math.isclose(a, b)`.

</details>

### Trap 8 — predict, then run

Many people expect: **True — it is the very same value**.

```python
missing_score = float('nan')
missing_score == missing_score
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `False`

`NaN` (a missing/invalid number) is defined to equal nothing, not even itself. Test it with `math.isnan(x)`.

</details>

### Trap 10 — predict, then run

Many people expect: **-3 (just chop off the decimal)**.

```python
net = -7
net // 2
```

<details>

### Trap 11 — predict, then run

Many people expect: **True — same digit**.

```python
cell = '5'   # a value read from a CSV
cell == 5
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `False`

Python does no automatic text/number conversion, so a CSV string and a number are simply unequal (no error). Convert first: `int(cell) == 5`.

</details>

### Trap 12 — predict, then run

Many people expect: **False (or maybe True)**.

```python
cell = '5'
5 > cell
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `TypeError: '>' not supported between instances of 'int' and 'str'`

Ordering a number against text raises TypeError — there is no sensible order. Convert first: `5 > int(cell)`.

</details>

<details>
<summary><strong>What really happens — and why</strong></summary>

### Trap 14 — predict, then run

Many people expect: **True — bools are ints**.

```python
flag = True
type(flag) is int
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `False`

`type()` is exact, and `flag`'s type is `bool`, not `int`. Use `isinstance` when you want subclass-aware checks.

</details>

### Trap 15 — predict, then run

Many people expect: **False — it's a bool, not an int**.

```python
flag = True
isinstance(flag, int)
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `True`

`isinstance` respects subclassing, and `bool` IS a kind of `int`.

</details>

### Trap 16 — predict, then run

Many people expect: **False — it's zero**.

```python
response = '0'   # what a respondent typed
bool(response)
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `True`

Any NON-EMPTY string is truthy, including '0' and 'False'. Only the empty string is falsy — so convert survey text before testing it.

</details>

### Trap 17 — predict, then run

Many people expect: **True — the list exists**.

```python
submissions = []
bool(submissions)
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `False`

Empty containers (`[]`, `{}`, `''`, `0`, `None`) are all falsy, which is why `if submissions:` reads as 'are there any?'.

</details>

### Trap 18 — predict, then run

Many people expect: **True — same number**.

```python
id_a = int('257')
id_b = int('257')   # two IDs parsed from text
id_a is id_b
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `False`

CPython pre-caches small ints (-5..256), so `is` accidentally looks True there; 257 built at runtime are separate objects. Compare values with `==`, never `is`.

</details>
