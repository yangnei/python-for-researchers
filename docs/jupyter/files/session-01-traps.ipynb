# Session 1 — Traps: predict, then run

Python's dynamic typing has sharp edges. For each cell below: **read it, predict the result out loud, then run it with Shift + Enter.** When the answer surprises you, that's the cell doing its job — expand *why* and change a value to test your new understanding.

**Tips:** press **Tab** to autocomplete a name, **Shift + Tab** for a function's help. Need a library? Run `%pip install <name>` in a cell (e.g. `%pip install pandas`) — in the browser (JupyterLite) it fetches a Pyodide build and lasts for the session.

### Trap 1 — predict, then run

Many people expect: **True — the lists are equal**.

```python
a = [1, 2]
b = [1, 2]
a is b
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `False`

`==` compares VALUES; `is` compares IDENTITY (the object in memory). Two equal-looking lists are still different objects. Use `==` for values and keep `is` for `None`.

</details>

### Trap 2 — predict, then run

Many people expect: **[1, 2] — we only touched a**.

```python
a = [1, 2]
b = a
a.append(3)
b
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `[1, 2, 3]`

`b = a` binds the SAME list, not a copy, so mutating through `a` shows up through `b`. Copy with `list(a)` or `a[:]`.

</details>

### Trap 3 — predict, then run

Many people expect: **False — they are different types**.

```python
True == 1
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `True`

`bool` is a subclass of `int`: `True == 1` and `False == 0`.

</details>

### Trap 4 — predict, then run

Many people expect: **TypeError — you can't add a bool to an int**.

```python
5 + True
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `6`

`True` acts as `1` (and `False` as `0`) in arithmetic, so `5 + True == 6`.

</details>

### Trap 5 — predict, then run

Many people expect: **an error, or 0**.

```python
flags = [True, False, True, True]
sum(flags)
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `3`

Summing booleans counts the `True`s — a handy way to count how many tests passed.

</details>

### Trap 6 — predict, then run

Many people expect: **0.3**.

```python
0.1 + 0.2
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `0.30000000000000004`

Floats are stored in binary, where 0.1 and 0.2 have no exact representation, so the tiny errors add up.

</details>

### Trap 7 — predict, then run

Many people expect: **True**.

```python
0.1 + 0.2 == 0.3
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `False`

Because `0.1 + 0.2` is `0.30000000000000004`. Never test floats with `==`; use `math.isclose(a, b)`.

</details>

### Trap 8 — predict, then run

Many people expect: **True — it is the very same value**.

```python
nan = float('nan')
nan == nan
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `False`

`NaN` is defined to be equal to nothing, not even itself. Test it with `math.isnan(x)`.

</details>

### Trap 9 — predict, then run

Many people expect: **3**.

```python
7 / 2
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `3.5`

`/` is always float (true) division. Use `//` for floor division.

</details>

### Trap 10 — predict, then run

Many people expect: **-3 (chop toward zero)**.

```python
-7 // 2
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `-4`

`//` floors toward negative infinity, not toward zero, so `-7 // 2 == -4`.

</details>

### Trap 11 — predict, then run

Many people expect: **True — same digit**.

```python
5 == '5'
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `False`

Python does no automatic number/text conversion, so a number and a string are simply unequal (no error).

</details>

### Trap 12 — predict, then run

Many people expect: **False (or maybe True)**.

```python
5 > '5'
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `TypeError: '>' not supported between instances of 'int' and 'str'`

Ordering a number against text raises TypeError — there is no sensible order. Convert first: `5 > int('5')`.

</details>

### Trap 13 — predict, then run

Many people expect: **True — same elements**.

```python
[1, 2] == (1, 2)
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `False`

A list never equals a tuple, whatever the contents.

</details>

### Trap 14 — predict, then run

Many people expect: **True — bools are ints**.

```python
type(True) is int
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `False`

`type()` is exact and `True`'s type is `bool`, not `int`. Use `isinstance` when you want subclass-aware checks.

</details>

### Trap 15 — predict, then run

Many people expect: **False — it's a bool, not an int**.

```python
isinstance(True, int)
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `True`

`isinstance` respects subclassing, and `bool` IS a kind of `int`.

</details>

### Trap 16 — predict, then run

Many people expect: **False — it's zero**.

```python
bool('0')
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `True`

Any NON-EMPTY string is truthy, including '0' and 'False'. Only the empty string is falsy.

</details>

### Trap 17 — predict, then run

Many people expect: **True — the list exists**.

```python
bool([])
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `False`

Empty containers (`[]`, `{}`, `''`, `0`, `None`) are all falsy.

</details>

### Trap 18 — predict, then run

Many people expect: **True — same number**.

```python
m = int('257')
p = int('257')
m is p
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `False`

CPython pre-caches small ints (-5..256), so `is` accidentally looks True there; 257 built at runtime are separate objects. Compare values with `==`, never `is`.

</details>
