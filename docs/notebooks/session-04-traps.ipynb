Python's dynamic typing has sharp edges. For each cell below: **read it, predict the result out loud, then run it with Shift + Enter.** When the answer surprises you, that's the cell doing its job — expand *why* and change a value to test your new understanding.

### Trap 1 — predict, then run

Many people expect: **3**.

```python
int('3.0')
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `ValueError: invalid literal for int() with base 10: '3.0'`

`int()` parses INTEGER text only; '3.0' is not a valid int literal. Use `int(float('3.0'))`.

</details>

### Trap 2 — predict, then run

Many people expect: **3**.

```python
round(2.5)
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `2`

Python uses banker's rounding (round half to EVEN): `round(2.5) == 2` but `round(3.5) == 4`. Watch this when summarising data.

</details>

### Trap 3 — predict, then run

Many people expect: **True**.

```python
0.1 + 0.2 + 0.3 == 0.6
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `False`

Float error accumulates when you total measurements: 0.1+0.2+0.3 is 0.6000000000000001. Compare sums with `math.isclose`, or round for display.

</details>

### Trap 4 — predict, then run

Many people expect: **ValueError**.

```python
int('1_000')
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `1000`

Python accepts underscores as digit separators, even inside `int()` text.

</details>
