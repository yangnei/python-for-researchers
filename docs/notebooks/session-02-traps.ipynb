### Trap 1 — predict, then run

Many people expect: **True or False**.

```python
typed_name = ''   # the user left the box blank
typed_name or 'Anonymous'
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `'Anonymous'`

`or` returns the first truthy OPERAND (else the last), and `and` the first falsy — not a bool. This is the default-value idiom.

</details>

### Trap 2 — predict, then run

Many people expect: **[1, 2, 3, 4, 5]**.

```python
weeks = range(1, 5)
list(weeks)
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `[1, 2, 3, 4]`

`range(start, stop)` stops BEFORE `stop`, so weeks 1–4 here.

</details>

### Trap 3 — predict, then run

Many people expect: **only the first student's first score changes**.

```python
gradebook = [[0] * 3] * 3   # 3 students x 3 assignments
gradebook[0][0] = 9
gradebook
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `[[9, 0, 0], [9, 0, 0], [9, 0, 0]]`

`[[0]*3]*3` makes three references to ONE inner row, so editing one edits all. Build with `[[0]*3 for _ in range(3)]`.

</details>

### Trap 4 — predict, then run

Many people expect: **True**.

```python
scores = [55, 92, 78]
all(s >= 60 for s in scores)
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `False`

`all()` is True only if EVERY item passes; 55 fails, so it's False.

</details>

### Trap 5 — predict, then run

Many people expect: **KeyError**.

```python
gpa = {'Ana': 3.9}
gpa.get('Ben')
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `None`

`.get()` returns `None` (or a default) for a missing key instead of raising — unlike `gpa['Ben']`.

</details>

### Trap 6 — predict, then run

Many people expect: **an error, or the same list**.

```python
ranking = ['Ana', 'Ben', 'Cara']
ranking[::-1]
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `['Cara', 'Ben', 'Ana']`

A slice with step -1 returns a reversed COPY — a common Python idiom.

</details>
