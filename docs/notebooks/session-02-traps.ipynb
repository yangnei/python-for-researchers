# Session 2 — Traps: predict, then run

Python's dynamic typing has sharp edges. For each cell below: **read it, predict the result out loud, then run it with Shift + Enter.** When the answer surprises you, that's the cell doing its job — expand *why* and change a value to test your new understanding.

**Tips:** press **Tab** to autocomplete a name, **Shift + Tab** for a function's help. Need a library? Run `%pip install <name>` in a cell (e.g. `%pip install pandas`) — in the browser (JupyterLite) it fetches a Pyodide build and lasts for the session.

### Trap 1 — predict, then run

Many people expect: **True or False**.

```python
5 and 0
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `0`

`and`/`or` return one of the OPERANDS, not a bool: `and` yields the first falsy value (else the last), `or` the first truthy.

</details>

### Trap 2 — predict, then run

Many people expect: **[1, 2, 3, 4, 5]**.

```python
list(range(1, 5))
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `[1, 2, 3, 4]`

`range(start, stop)` stops BEFORE `stop`.

</details>

### Trap 3 — predict, then run

Many people expect: **[[9, 0, 0], [0, 0, 0], [0, 0, 0]]**.

```python
grid = [[0] * 3] * 3
grid[0][0] = 9
grid
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `[[9, 0, 0], [9, 0, 0], [9, 0, 0]]`

`[[0]*3]*3` makes three references to ONE inner row, so editing one edits all. Build with `[[0]*3 for _ in range(3)]`.

</details>

### Trap 4 — predict, then run

Many people expect: **True**.

```python
scores = [55, 92, 78]
all(s >= 60 for s in scores)
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `False`

`all()` is True only if EVERY item passes; 55 fails, so it's False.

</details>

### Trap 5 — predict, then run

Many people expect: **KeyError**.

```python
d = {'Ana': 91}
d.get('Ben')
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `None`

`.get()` returns `None` (or a default) for a missing key instead of raising — unlike `d['Ben']`.

</details>

### Trap 6 — predict, then run

Many people expect: **an error, or the same list**.

```python
nums = [1, 2, 3, 4]
nums[::-1]
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `[4, 3, 2, 1]`

A slice with step -1 returns a reversed COPY — a common Python idiom.

</details>
