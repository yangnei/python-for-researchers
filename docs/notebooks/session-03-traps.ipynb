### Trap 1 — predict, then run

Many people expect: **['banana'] — a fresh empty bag each call**.

```python
def add(item, bag=[]):
    bag.append(item)
    return bag
add('apple')
add('banana')
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `['apple', 'banana']`

A default value is created ONCE, at def time, so the same list persists across calls. Use `bag=None` then `bag = bag or []`.

</details>

### Trap 2 — predict, then run

Many people expect: **False — it clearly produced 42**.

```python
def show(x):
    print(x)
show(42) is None
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `True`

Printing is not returning. A function with no `return` gives `None`.

</details>

### Trap 3 — predict, then run

Many people expect: **[0, 1, 2]**.

```python
funcs = [lambda: i for i in range(3)]
[f() for f in funcs]
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `[2, 2, 2]`

Closures capture the VARIABLE `i`, not its value; by call time the loop has left `i = 2`. Fix by binding it: `lambda i=i: i`.

</details>

### Trap 4 — predict, then run

Many people expect: **runs forever, or 0**.

```python
def f(n):
    return f(n - 1)
f(3)
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `RecursionError: maximum recursion depth exceeded`

Every recursive function needs a base case. Without one Python stops at its recursion limit (~1000 deep) with RecursionError.

</details>
