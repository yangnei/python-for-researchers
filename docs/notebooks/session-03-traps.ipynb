### Trap 1 — predict, then run

Many people expect: **['Ben'] — a fresh empty roster each call**.

```python
def enroll(name, roster=[]):
    roster.append(name)
    return roster
enroll('Ana')
enroll('Ben')
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `['Ana', 'Ben']`

A default value is created ONCE, at def time, so the same list persists across calls. Use `roster=None` then `roster = roster or []`.

</details>

### Trap 2 — predict, then run

Many people expect: **False — it clearly produced 91**.

```python
def show_score(s):
    print(s)
show_score(91) is None
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `True`

Printing is not returning. A function with no `return` gives `None`.

</details>

### Trap 3 — predict, then run

Many people expect: **[0, 1, 2]**.

```python
makers = [lambda: week for week in range(3)]
[make() for make in makers]
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `[2, 2, 2]`

Closures capture the VARIABLE `week`, not its value; by call time the loop has left `week = 2`. Fix by binding it: `lambda week=week: week`.

</details>

### Trap 4 — predict, then run

Many people expect: **runs forever, or 0**.

```python
def countdown(n):
    return countdown(n - 1)   # no base case!
countdown(3)
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `RecursionError: maximum recursion depth exceeded`

Every recursive function needs a base case. Without one Python stops at its recursion limit (~1000 deep) with RecursionError.

</details>
