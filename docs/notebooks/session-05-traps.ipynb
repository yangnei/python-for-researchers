### Trap 1 — predict, then run

Many people expect: **the same GPAs again**.

```python
gpas = (s for s in [3.9, 1.8, 3.2])
first_pass = list(gpas)
list(gpas)
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `[]`

A generator is one-shot: after the first full pass it is exhausted. Rebuild it, or store a list if you need it twice.

</details>

### Trap 2 — predict, then run

Many people expect: **False — two different objects**.

```python
from dataclasses import dataclass

@dataclass
class Grade:
    course: str
    score: int

g1 = Grade('ED101', 91)
g2 = Grade('ED101', 91)
g1 == g2
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `True`

`@dataclass` writes an `__eq__` that compares fields, so equal-valued grades are `==` (even though `g1 is g2` is False).

</details>

### Trap 3 — predict, then run

Many people expect: **'ED101'**.

```python
import re
m = re.match(r'\[(.+)\]', '[ED101][ED102]')
m.group(1)
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `'ED101][ED102'`

`+` is greedy — it grabs as much as it can. Use `+?` for the shortest match: `r'\[(.+?)\]'`.

</details>

### Trap 4 — predict, then run

Many people expect: **True — there's a number in there**.

```python
import re
bool(re.match(r'\d+', 'cohort2026'))
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `False`

`re.match` anchors at the START of the string. Use `re.search` to find a match anywhere.

</details>

### Trap 5 — predict, then run

Many people expect: **[] — math is a different course**.

```python
class Course:
    students = []
    def enroll(self, name):
        self.students.append(name)

art = Course()
math = Course()
art.enroll('Ana')
math.students
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `['Ana']`

`students` is a CLASS variable shared by every instance. Give each its own in `__init__`: `self.students = []`.

</details>
